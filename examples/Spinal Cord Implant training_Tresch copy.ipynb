```python
import cv2
import csv
from pathlib import Path

def attempttomakefolder(foldername, recursive=False):
    """ Attempts to create a folder with specified name. Does nothing if it already exists. """
    try:
        os.path.isdir(foldername)
    except TypeError:
        foldername = os.fspath(
            foldername
        )

    if os.path.isdir(foldername):
        print(foldername, " already exists!")
    else:
        if recursive:
            os.makedirs(foldername)
        else:
            os.mkdir(foldername)


def Getlistofvideos(videos, videotype):
    if [os.path.isdir(i) for i in videos] == [True]:  # checks if input is a directory
        from random import sample

        videofolder = videos[0]

        os.chdir(videofolder)
        videolist = [
            os.path.join(videofolder, fn)
            for fn in os.listdir(os.curdir)
            if os.path.isfile(fn)
            and fn.endswith(videotype)
            and "_labeled." not in fn
            and "_full." not in fn
        ]

        Videos = sample(
            videolist, len(videolist)
        )
    else:
        if isinstance(videos, str):
            if (
                os.path.isfile(videos)
                and "_labeled." not in videos
                and "_full." not in videos
            ):  # #or just one direct path!
                Videos = [videos]
            else:
                Videos = []
        else:
            Videos = [
                v
                for v in videos
                if os.path.isfile(v) and "_labeled." not in v and "_full." not in v
            ]
    return Videos

fields = ['Channel', 'Amplitude', 'Pulse Width', 'Frequency', 'Resolution'] 
for folder in videofile_pathlist:
    tmpfolder = os.path.join(folder, "capture-frames")
    dataFile = os.path.join(folder, "capture-frames", "video-data.csv")
    print(dataFile)
    attempttomakefolder(tmpfolder, recursive=True)

    Videos = Getlistofvideos([folder], '.avi')
    with open(dataFile, 'w') as csv_file:
        csvwriter = csv.writer(csv_file)
        csvwriter.writerow(fields)

        for video in Videos:
            videofolder = str(Path(video).parents[0])
            videoName = str(Path(video).stem)

            framepath = (os.path.join(tmpfolder, videoName + "." + "bmp"))
            csvwriter.writerows([[videoName, 0]])

            vidcap = cv2.VideoCapture(video)
            success,image = vidcap.read()

            cv2.imwrite(framepath, image) 

```

```python
import re
import os
from datetime import datetime
import pandas as pd
import numpy as np
import cv2
from pathlib import Path

expressionChannel = "(?<=C)[0-9]+";
expressionAmplitude = "(?<=AMP)[0-9]+";
expressionFrequency = "(?<=[0-9]P)[0-9]+";
expressionPulseWidth = "(?<=PW)[0-9]+";

fields = ['Channel', 'Amplitude', 'Frequency', 'Pulse Width', 'Recorded date', 'Resolution']
folder = '/content/drive/MyDrive/Rat/SCI tests'
dataFile = os.path.join(folder, "video-data.csv")
try:
    df = pd.read_csv(dataFile, usecols=fields)
    df['Recorded date'] = pd.to_datetime(df['Recorded date'], dayfirst=True)
    print('reading existing file')
except:
    df = pd.DataFrame(columns=fields)

for folder in videofile_pathlist:
    
    Videos = Getlistofvideos([folder], '.avi')
    for video in Videos:
        videoName = str(Path(video).stem)
        videoNameSplit = videoName.split('_')
        print(videoName)
        for i in range(1, 5):
            try:
                recordingDate = videoNameSplit[i]
                recordingDate = str(datetime.strptime(recordingDate, '%Y-%m-%d').date())
                print("Recoded date is " + recordingDate)
                break
            except:
                pass

        channelData = re.findall(expressionChannel, videoName)
        if channelData:
            print("Channel =" + " " + channelData[0])
            channelData = float(channelData[0])
        else:
            channelData = -1

        amplitudeData = re.findall(expressionAmplitude, videoName)
        if amplitudeData:
            print("Amplitude =" + " " + amplitudeData[0])
            amplitudeData = float(amplitudeData[0])
        else:
            amplitudeData = -1

        frequencyData = re.findall(expressionFrequency, videoName)
        if frequencyData:
            print("Frequency =" + " " + frequencyData[0])
            frequencyData = float(frequencyData[0])
        else:
            frequencyData = -1

        pulsewidthData= re.findall(expressionPulseWidth, videoName)
        if pulsewidthData:
            print("Pulse width =" + " " + pulsewidthData[0])
            pulsewidthData = float(pulsewidthData[0])
        else:
            pulsewidthData = -1


        check_duplicate = (((df['Channel']) == channelData) & \
                           ((df['Amplitude']) == amplitudeData) & \
                           ((df['Frequency']) == frequencyData) & \
                           ((df['Pulse Width']) == pulsewidthData) & \
                           (df['Recorded date'] == recordingDate)).any()

        if not check_duplicate:
            df.loc[len(df)] = {
                'Channel': channelData,
                'Amplitude': amplitudeData,
                'Frequency': frequencyData,
                'Pulse Width': pulsewidthData,
                'Recorded date': pd.to_datetime(recordingDate)
            }
            pass
        else:
            print("Duplicate found!")
    
        print("")

df['Recorded date'] = pd.to_datetime(df['Recorded date'], dayfirst=True)
df.sort_values(by='Recorded date', inplace=True, ascending=False)
df = df.reset_index(drop=True)
df.fillna(-1)
df.to_csv(path_or_buf=dataFile, float_format='%.1f')
```
